###Importing Libraries

```python
from google.colab import files
import pandas as pd 

from sklearn.ensemble import RandomForestClassifier

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

###Data loading

```python
uploaded = files.upload()

train_data = pd.read_csv("train.csv")

uploaded = files.upload()

test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

###Following the given Instructions

```python
sur_women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_sur_women = sum(sur_women)/len(sur_women)

print("% of women who survived:", rate_sur_women)
```

```python
sur_men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_sur_men = sum(sur_men)/len(sur_men)

print("% of men who survived:", rate_sur_men)
```

```python
y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
X.head()
```

```python
X_test.head()
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
output.Survived.sum()
```

###My Contribution

####Train Data pre-processing

```python
print("Train Data:", train_data.shape)
print("Test Data:", test_data.shape)
```

```python
process_train = train_data
```

```python
#try to fill up missing values
process_train.isnull().sum()
```

```python
#Filling average of age values in the place of missing value
process_train['Age'].fillna(process_train['Age'].mean(),inplace=True)
```

```python
print("Rate of missing values in Cabin coloum:", (process_train.Cabin.isnull().sum()/process_train.shape[0]))
```

```python
#Dropping 'Cabin' due to lot of missing values
process_train = process_train.drop('Cabin', axis=1)
```

```python
process_train["Embarked"].value_counts()
```

```python
#Filling missing values in 'Embarked' with most frequently occuring value
process_train['Embarked'].fillna('S',inplace=True)
```

```python
#Dropping 'PassengerID', 'Name', 'Ticket', 'Fare' because they are not useful
process_train = process_train.drop(['PassengerId', 'Name', 'Ticket', 'Fare'], axis=1)
```

```python
#Dropping 'Survived' because it is dependent variable
process_train = process_train.drop('Survived', axis=1)
```

```python
process_train.head()
```

```python
train_cols =  list(process_train)
print(train_cols)
```

####Model Training using Voting Classifier 


```python
est = []
est.append(('LR', LogisticRegression(max_iter = 200)))
est.append(('RFC', RandomForestClassifier(n_estimators=900, max_depth=7, random_state=1)))
est.append(('KNR', KNeighborsClassifier(n_neighbors=4)))
est.append(('GNB', GaussianNB()))
est.append(('SVC', SVC()))
```

```python
vc_model = VotingClassifier( estimators=est, voting='hard')
vc_model.fit(X_tr, y_tr)
predictions_vc = vc_model.predict(X_ts)
```

```python
output_vc = pd.DataFrame({'PassengerId': process_test.PassengerId, 'Survived': predictions_vc})
output_vc.to_csv('submission_vc.csv', index=False)
print("Your submission was successfully saved!")
```
